```python
import cv2
import os
import yaml
```

```python
with open("config.yaml", "r") as file:
    config = yaml.safe_load(file)["vid_to_frame"]
```

```python
video_path = config["input"] + config["file_name"]
output_dir = config["output"]
os.makedirs(output_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
frame_rate = config["parameters"]["frame_rate"]
frame_count = config["parameters"]["frame_count"]
saved_count = config["parameters"]["saved_count"]
```

```python
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break
    
    if frame_count % frame_rate == 0:
        frame_filename = f"{output_dir}/frame_{saved_count:04d}.jpg"
        cv2.imwrite(frame_filename, frame)
        saved_count += 1

    frame_count += 1
```

```python
cap.release()
print(f"Extracted {saved_count} frames.")
```
