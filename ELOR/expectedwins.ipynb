```python
import pandas as pd 
import numpy as np 
from tqdm import tqdm
```

```python
df = pd.read_csv("Datasets/ELO.csv")
```

```python
teams = list(set(df.Home_Team) | set(df.Away_Team))
years = list(df.Year.unique())

results = []
for team in teams:
    df_team_1 = df[(df.Home_Team == team) | (df.Away_Team == team)]
    for year in years:
        df_team = df_team_1[(df_team_1.Year == year)]
        expected = 0
        real = 0
        for x in range(len(df_team)):
            if df_team.Home_Team.iloc[x] == team:
                expected += df_team.Prob_Home.iloc[x]
                real += df_team.Home_Away_Draw.iloc[x]
            else:
                expected += df_team.Prob_Home.iloc[x]
                real += abs(df_team.Home_Away_Draw.iloc[x]-1)
        results.append((year,team,expected,real,real-expected))

df1 = pd.DataFrame(results, columns=["Year","Team","xWins","Wins","Performance"])
df1 = df1[df1.xWins > 0]
```

```python
df1.to_csv("ELO/All_time_wins_year.csv")
```
